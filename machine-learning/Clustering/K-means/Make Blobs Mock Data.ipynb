# K Means Clustering with Mock Data


Data: Used the make_blobs function in Python to generate mock data

Reference: [Make Blobs](http://scikit-learn.org/stable/modules/generated/sklearn.datasets.make_blobs.html)

## Import Libraries

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Create Data

Create 200 samples using make_blobs.

```python
from sklearn.datasets import make_blobs
```

```python
data = make_blobs(n_samples=200, n_features=2, 
                           centers=4, cluster_std=1.8,random_state=101)
```

## Visualize Data

```python
plt.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

## Creating the Clusters

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=4)
```

```python
kmeans.fit(data[0])
```

```python
kmeans.cluster_centers_
```

```python
kmeans.labels_
```

## Comparison

Although we will compare our clustering results with original cluster but in real-world examples, since K-Means is unsupervised algorithm, there is no original data to compare with.

```python
f, (ax1, ax2) = plt.subplots(1, 2, sharey=True,figsize=(10,6))
ax1.set_title('K Means')
ax1.scatter(data[0][:,0],data[0][:,1],c=kmeans.labels_,cmap='rainbow')
ax2.set_title("Original")
ax2.scatter(data[0][:,0],data[0][:,1],c=data[1],cmap='rainbow')
```

Please note the colors have no relevance between the two charts.
